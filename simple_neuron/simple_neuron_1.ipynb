```python
ve = 0.0
h = 0.001
y1 = 20
y2 = 20
inp = 0.02
threshold = 10
reset = -3

s1_2=0.0
syn_cons = 0.1
syn_peak = 50.0

y_values = []

for i in range(10000):
    y1 = y1 + h*(ve-y1) + inp
    if y1 > threshold:
        y1 = reset
        s1_2 = syn_peak
        
    s1_2 = syn_cons * s1_2
        
    y2 = y2 + h*(ve-y2) + s1_2
    if y2 > threshold:
        y2 = reset    
        
    y_values.append([y1, y2, s1_2])
```

```python
plt.plot([y[0] for y in y_values], label='y1')
plt.plot([y[1] for y in y_values], label='y2')
plt.plot([y[2] for y in y_values], label='s12')
plt.legend()
plt.show()
```
